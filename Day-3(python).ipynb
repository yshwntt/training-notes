A string is text data like names, messages, or sentences. In Python, a string is a sequence, so every character has a position (index).

In Jupyter, writing a string alone shows output, but normally we use print():

```python
print("Hello World")
```

```python
len("Hello")
```

Index starts from 0:

```python
s = "Hello"
s[0]   # H
s[1]   # e
```

Slicing gives part of a string

```python
s = "Hello World"
s[:5]     # Hello
s[6:]     # World
s[:]      # full string
```

Negative index counts from the end:

```python
s[-1]   # last character
s[:-1]  # everything except last
```

Step slicing

```python
s[::2]   # every 2nd character
s[::-1]  # reverse string
```

Immutable

```python
# s[0] = 'X'  # error
s = "Hello"
s = "Y" + s[1:]   # Yello
```

concatenation + repetition

```python
"a" + "b"     # ab
"z" * 5       # zzzzz
```

```python
s = "Hello world"
s.upper()      # HELLO WORLD
s.lower()      # hello world
s.split()      # ['Hello', 'world']
```

split('x') removes the separator
partition('x') keeps the separator and returns a tuple

```python
"hello4".split("e")        # ['h', 'llo4']
"hellloepp".partition("e") # ('h', 'e', 'llloepp')
```

count and find

```python
s.count("o")   # how many times
s.find("o")    # first index of 'o'
```

```python
"hello4".isalnum()   # letters+numbers only
"Ruchik".isalpha()   # only letters
"yash".islower()     # all lowercase
" ".isspace()        # only spaces
"Hello".istitle()    # Title Case
"HELLO".isupper()    # all uppercase
"hello4".endswith("4")
```

List = collection of items (can store different data types).

```python
my_list = [1, 2, 3]
mix = ["A", 10, 2.5, "x"]
len(mix)
```

Indexing and slicing (same as strings)

```python
my_list = ["one", "two", "three", 4, 5]
my_list[0]     # one
my_list[1:]    # from index 1 to end
my_list[:5:2]  # step slicing
```

list concatenation and repetition

```python
my_list + ["new"]
my_list * 2
```

append() adds ONE item (even if itâ€™s a list)
extend() adds elements one by one

```python
x = [1, 2, 3]
x.append([4, 5])   # [1, 2, 3, [4, 5]]

x = [1, 2, 3]
x.extend([4, 5])   # [1, 2, 3, 4, 5]
```

```python
l = [10, 20, 30]
l.pop()     # removes 30
l.pop(0)    # removes 10 #pop 
```

```python
l = [1, 2, 3, 2]
l.remove(2)   # removes first 2 #remove
```

```python
new_list = ['a','e','x','b','c','1','2']
new_list.sort()     # sorts (numbers as strings come first)
new_list.reverse()  # reverses in place
```
