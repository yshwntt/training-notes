Functions are used to group code so you can run it many times without rewriting.

```python
def greeting(name):
    print("Hello", name)

greeting("Yash")
greeting(90)   # python allows any datatype
```

print() only displays output
return sends value back so we can store it in variable

```python
def add_num(a, b):
    return a + b

result = add_num(5, 6)
print(result)
print(result + 5)
```

```python
def add_num(a, b):
    print(a + b)

x = add_num(5, 6)   # x becomes None because no return
print(x)
```

loop + break

```python
def is_prime(num):
    if num <= 1:
        print("not prime")
        return

    for i in range(2, num):
        if num % i == 0:
            print("not prime")
            break
    else:
        print("prime")

is_prime(9)
is_prime(7)
```

Generator = function that gives values one by one using yield.
return stops function fully
yield pauses and remembers where it stopped

```python
def gencubes(n):
    for i in range(n):
        yield i**3

for x in gencubes(5):
    print(x)
```

Iterable = you can loop over it (string/list)
Iterator = gives next value using next()

```python
def simple_gen():
    for x in range(3, 6):
        yield x

g = simple_gen()
print(next(g))
print(next(g))
print(next(g))
#next(g)  -> StopIteration after values finish
```

map(function, iterable) applies function to each element.

```python
lst = [1, 2, 3, 4]
new_lst = []

for x in lst:
    new_lst.append(x + 1)

print(new_lst)
```

using map + lambda

```python
lst = [1, 2, 3, 4]
print(list(map(lambda x: x + 1, lst)))
```

reduce() keeps combining items until one value left.

```python
from functools import reduce

names = ["ash", "grey", "worm"]
result = reduce(lambda a, b: a + b, names)
print(result)
```

filter(function, iterable) keeps only values where function returns True.

```python
lst = [1,2,3,4,5,6,7,8]
evens = list(filter(lambda x: x % 2 == 0, lst))
print(evens)
```
