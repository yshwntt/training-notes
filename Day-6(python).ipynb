try & except

```python
try:
    x = 10 / 0
except:
    print("Error happened")
```

Specific exception

```python
try:
    num = int("abc")
except ValueError:
    print("Invalid number")
```

Multiple except

```python
try:
    a = 10
    b = int(input("Enter number: "))
    print(a / b)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Enter valid number")
```

else block

```python
try:
    result = 10 / 2
except ZeroDivisionError:
    print("Error")
else:
    print("No error, result is", result)
```

finally block

```python
try:
    print("Inside try")
except:
    print("Error")
finally:
    print("This always runs")
```

raise keyword

```python
def check_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    else:
        print("Valid age")

check_age(5)
# check_age(-1)
```

Custom exception

```python
class NegativeNumberError(Exception):
    pass

def check_number(num):
    if num < 0:
        raise NegativeNumberError("Negative not allowed")
    else:
        print("Valid number")

check_number(10)
```
