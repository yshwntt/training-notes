Class variable and instance variable

```python
class Student:
    school = "ABC School"   # Class variable

    def __init__(self, name):
        self.name = name    # Instance variable at the instance level shared variables

s1 = Student("Alice") # only accesible 
s2 = Student("Bob")

print(s1.school)   # ABC School
print(s2.school)   # ABC School
print(s1.name)
print(s2.name)

Student.school = "XYZ School"   # Change class variable at global level

print(s1.school)   # XYZ School
print(s2.school)   # XYZ School
```

```python
s1.school = "s1 Own School"
print(s1.school)   # Own School (instance var now)
print(s2.school)   # XYZ School
```

Instance Method can access class variable but not the other way(i.e instance variable cannot be used outside)

```python
class MyClass:
    class_var = 100   # Class variable

    def __init__(self, value):
        self.value = value   # Instance variable

    def show(self):   # Instance method
        print(f"Instance value: {self.value}")
        print(f"Class variable before: {MyClass.class_var}")

        # Modifying both instance and class variable
        self.value += 10
        MyClass.class_var -= 5

        print(f"Instance value after: {self.value}") # second instance method
        print(f"Class variable after: {MyClass.class_var}")

# Create an object and call the instance method
obj = MyClass(15)
obj.show()
```

class method

```python
class MyClass:
    count = 0   # class variable
    rk = 0

    def __init__(self):
        MyClass.rk += 1 #increment inst +1

    @classmethod
    def get_instance_count(cls):
        return cls.rk


print(MyClass.get_instance_count())
a = MyClass() # 1st instance
b = MyClass() # 2nd i
c = MyClass() # 3rd i
print(MyClass.get_instance_count())
```

```python
class MyClass:

    def __init__(self, value):
        self.value = value   # INSTANCE VARIABLE (belongs to each object separately)

    @classmethod
    def show_value(cls):
        # cls refers to the class itself, NOT the object
        # Class methods can only directly access class variables
        # They CANNOT directly access instance variables like self.value
        print(cls.value)     # This will cause an error because 'value' is not a class variable


obj = MyClass(10)          # Creating object with instance value = 10
MyClass.show_value()       # Calling class method using class name
```

static methods

```python
class MyClass:

    def __init__(self, value):
        self.value = value   # Instance variable (belongs to each object)

    @staticmethod
    def show_value(obj):
        # Static method does NOT receive self or cls automatically
        # It behaves like a normal function inside a class
        # We must manually pass the object if we want to access instance variables
        print(obj.value)     # Accessing the instance variable through the passed object


a = MyClass(42)             # Creating an object with value = 42
MyClass.show_value(a)       # Passing object manually to static method
```

Decorator

```python
def exp(fun):                      # Decorator function (takes another function as argument)
    def wrapper():                 # Wrapper function (adds extra behavior)
        print("Before the function runs")
        fun()                      # Calling the original function
        print("After the function runs")
    return wrapper                 # Returning the modified function


# def say_hello():
#     print("Hello: RAM")

# Apply the decorator manually
# decorated_func = rk(say_hello)
# decorated_func()

@exp                               # Using decorator syntax (same as: say_hello = exp(say_hello))
def say_hello():
    print("Hello!")

say_hello()                        # Calls wrapper(), not the original function directly
```

How long should a class should be - should have one responsibility have one reason to change for software maintainance

```python
class Report:
    def __init__(self, title, content):
        self.title = title          # Instance variable storing report title
        self.content = content      # Instance variable storing report content

    def generate(self):
        # Returns formatted report as a single string
        # \n adds a new line between title and content
        return f"{self.title}\n{self.content}"


class ReportSaver:
    def save_to_file(self, report, filename):
        # Opens the file in write mode ("w")
        # If file doesn't exist → it creates it
        # If file exists → it overwrites it
        with open(filename, "w") as f:
            # Calls report.generate() to get formatted text
            # Writes the text into the file
            f.write(report.generate())
```
