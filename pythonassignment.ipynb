Palindrome

```python
text = "madam"

reverse = ""

for ch in text:
    reverse = ch + reverse   # building reverse string

if text == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```

Factorial

```python
num = 5
fact = 1

for i in range(1, num + 1):
    fact = fact * i

print("Factorial is:", fact)
```

Prime number

```python
num = 7
is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print("Prime")
else:
    print("Not Prime")
```

Count vowels in string

```python
text = "education"
count = 0

for ch in text:
    if ch in "aeiou":
        count += 1

print("Vowel count:", count)
```

Fibonacci series (first n numbers)

```python
n = 6

a = 0
b = 1

for i in range(n):
    print(a)
    c = a + b
    a = b
    b = c
```
